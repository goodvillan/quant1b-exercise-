## Part B

* In linear regresssion we can only use numerical variables. 
* we can convert the categorical columns into numerical ones using one hot encoding to use both categorical and numerical data. 


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df.corr()['positive_mental_health_scale']
```

*  income has the highest correlation with positive_mental_health_scale (70%), followed by exercise_hours_weekly 40% 

X contains the independent variables:'exercise_hours_weekly', 'income'

```python
x = df[['exercise_hours_weekly', 'income']]
```

```python
y = df['positive_mental_health_scale']
```

* Separate the data into train and test part

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

#### Prepare Model

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

The above code fits the linear regression model on the training data.

```python
predictions = model.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.show()
```

```python
r_sq = model.score(X_test, y_test)
print('coefficient of determination:', round(r_sq,1))
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, predictions)
```

```python
print(f"R2 Score is {round(r2,1)}")
```
